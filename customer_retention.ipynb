# Examining the data:

```python
#creating a pandas dataframe that contains the information of the dataset
df=pd.read_excel("customer_retention_dataset.xlsx")
df.head()
```

We can observe that the dataset is having almost categorical data. And the headers are having an indexes so let us remove those digits from the header's strings.

```python
from string import digits
#removing tabspace
df.columns=df.columns.str.replace("\t"," ")

#removing digits
remove_digits=str.maketrans('','',digits)
df.columns=df.columns.str.translate(remove_digits)

df.columns=df.columns.str.strip()
```

Now let us analyse the majority of loyal customers shop from which online retailers.

```python
dict={'31-40 times':35,'41 times and above':45,'Less than 10 times':5,'11-20 times':15,'21-30 times':25,'42 times and above':45}
df['Average times made an online purchase']=df['How many times you have made an online purchase in the past  year?'].replace(dict)
```

```python
plt.figure(figsize=(20,8))
sb.violinplot(x='From the following, tick any (or all) of the online retailers you have shopped from;',y='Average times made an online purchase',hue='Gaining access to loyalty programs is a benefit of shopping online',data=df)
plt.xticks(rotation=45)
```

People who are shopping from amazon and paytm are most loyal. Flipkart and snapdeal also have this but people who shop almost everywhere disagree with this point. 

```python
#checking loyalty of the customers
bad=['Longer time to get logged in (promotion, sales period)',
       'Longer time in displaying graphics and photos (promotion, sales period)',
       'Late declaration of price (promotion, sales period)',
       'Longer page loading time (promotion, sales period)',
       'Limited mode of payment on most products (promotion, sales period)',
       'Longer delivery period', 'Change in website/Application design',
       'Frequent disruption when moving from one page to another']    
```

```python
for i in bad:
    plt.figure(figsize=(15,6))
    sb.countplot(df[i],hue=df['Which of the Indian online retailer would you recommend to a friend?'])
    plt.xticks(rotation=45)
    plt.figure()
```

Customers seem to be more loyal to amazon, fipkart and paytm as even though many of them have given negative remarks about them still they would recommend these platforms to their friend.

```python
retailer=df.groupby('From the following, tick any (or all) of the online retailers you have shopped from;')['Average times made an online purchase'].count().reset_index()
retailer.rename(columns={'Average times made an online purchase':'Count'},inplace=True)
retailer_sorted=retailer.sort_values('Count',ascending=False)
plt.figure(figsize=(10,5))
plt.plot(retailer_sorted['From the following, tick any (or all) of the online retailers you have shopped from;'][:10],retailer_sorted['Count'][:10],
        linestyle='-',color='purple',marker='o')
plt.title("Best-selling retailers",fontsize='15')
plt.xlabel('Retailer name',fontsize='15')
plt.ylabel('Number of Products Sold',fontsize='15')
plt.show()
```

We can see that most of the products are sold in Amazon and flipkart as of this analysis they are the best-selling retailers.
